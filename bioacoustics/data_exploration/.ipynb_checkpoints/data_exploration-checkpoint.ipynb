```python
import xarray as xr
import numpy as np
import s3fs
import subprocess
```

# Download dataset

Download file with bash 

```python
path = "s3-ap-southeast-2.amazonaws.com/imos-data/IMOS/SOOP/SOOP-BA/VLMJ_Investigator/Investigator_20160412-20160414/IMOS_SOOP-BA_AE_20160412T031347Z_VLMJ_FV02_Investigator-EK60-18_END-20160414T043251Z_C-20210622T061448Z.nc"
bashCmd = ["wget", path]
process = subprocess.Popen(bashCmd, stdout=subprocess.PIPE)
output, error = process.communicate()
```

Create xarray object

```python
name = "IMOS_SOOP-BA_AE_20160412T031347Z_VLMJ_FV02_Investigator-EK60-18_END-20160414T043251Z_C-20210622T061448Z.nc"
ds = xr.open_dataset(name)
ds
```

```python
ds["DEPTH"] = -ds["DEPTH"]
```

```python
# if one channel
fig = (10*np.log10(ds['Sv'])).plot(x='TIME', size=6, aspect=4, cmap="jet")

# if multiple channels
# fig = (10*np.log10(ds['Sv'])).plot(x='TIME', size=6, aspect=4, row='CHANNEL')
# or 
# fig = (10*np.log10(ds['Sv'].isel(CHANNEL=0))).plot(x='TIME', size=6, aspect=4, cmap="jet")
```

```python
fig.figure.savefig('base_18kHz.png')
```

```python
diff = ds['Sv'] - ds['Sv'].mean()
```
